```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
ticker = "AAPL"
df = yf.download(ticker, start="2021-01-01", end="2026-03-12")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['Returns'] = df['Close'].pct_change()
```

```python
df['Rolling_Vol'] = df['Returns'].rolling(window=21).std() * np.sqrt(252)
```

```python
df['Target_Vol'] = df['Rolling_Vol'].shift(-5)
```

```python
df.dropna(inplace=True)
```
